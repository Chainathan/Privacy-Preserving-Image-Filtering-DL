```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from datasets import load_dataset

ds = load_dataset("HosseinGT/wider_face_background")
```

```python
# Visualize the first 5 examples
for i in range(5):
    image = ds["train"][i]["image"]
    boxes = ds["train"][i]["faces"]["bbox"]
    plt.imshow(image)
    for box in boxes:
        x, y, w, h = box
        plt.gca().add_patch(plt.Rectangle((x, y), w, h, linewidth=1, edgecolor="r", facecolor="none"))
    plt.show()
```
